```python
import pandas as pd
import os 
import numpy as np
import tensorflow as tf
import cv2 
import matplotlib.pyplot as plt
from os import path
```

```python
def xywh2minmax(boxes):
    cx, cy, w, h = [boxes[..., i] for i in range(4)]
    xmin = cx - w*0.5
    ymin = cy - h*0.5
    xmax = cx + w*0.5
    ymax = cy + h*0.5
    new_boxes = np.stack([xmin, ymin, xmax, ymax], axis=-1)
    return np.concatenate([new_boxes, boxes[..., 4:]], axis=-1)
```

```python
base_path = 'Saffron_Dataset/Labeled/'
csv_files = [i for i in os.listdir(base_path) if i.endswith('csv')]
```

```python
csv_files = csv_files[:14]
```

```python
for csv in csv_files:
    
    data = pd.read_csv(base_path+csv,header=None,names=['cx','cy','angle'])
    data['w'] = 30
    data['h'] = 30 
    angle = data['angle']
    data.drop('angle',axis=1,inplace=True)
    boxes = xywh2minmax(np.array(data)).astype(np.int64)
    out = pd.DataFrame([(*boxes)],columns=['xmin','ymin','xmax','ymax'])
    out['class'] = np.array(np.ceil(abs(np.cos(angle * (np.pi/180)))*10),dtype=np.int64)
    out.to_csv(base_path+'box_'+csv,index=False,columns=None)
```

```python
def load_saffran_dataset(dataroot='./data/Saffron_Dataset/Labeled/'):
    img_paths, bboxes = [], []
    files = os.listdir(dataroot)
    jpegFiles = [i for i in files if i.endswith('.jpg')]

    for image_name in jpegFiles:
        name = image_name.split('.')[0]
        csv_name = 'box_{}.csv'.format(name)
        box = np.array(pd.read_csv(dataroot+csv_name))
        img_paths.append(path.join(dataroot+image_name))
        bboxes.append(box)
    return np.array(img_paths), np.array(bboxes)
```
